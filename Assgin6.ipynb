```python
import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter a value for M: "))

x = np.linspace(-10, 10, 400)
y1 = M * x**2
y2 = M * np.sin(x)

plt.plot(x, y1, color='blue', linestyle='-', linewidth=2)
plt.plot(x, y2, color='red', linestyle='--', linewidth=2)

plt.title('Mathematical Functions')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
roll_number = 102316060
np.random.seed(roll_number)
data = np.random.randn(50)
plt.subplot(2, 2, 1)
plt.plot(np.cumsum(data), label='Cumulative Sum', color='blue')
plt.title('Cumulative Sum')
plt.xlabel('Index')
plt.ylabel('Cumulative Sum')
plt.grid(True)
plt.subplot(2, 2, 2)
plt.scatter(range(50), data, color='green')
plt.title('Scatter Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

url = 'https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv'

data = pd.read_csv(url)
print(data.columns)
print(data.head())

plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
sns.lineplot(x='month_number', y='total_profit', data=data, marker='o', color='purple')
plt.title('Total Profit for Each Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.subplot(2, 2, 2)
sns.lineplot(data=data[['facecream', 'facewash', 'toothpaste', 'bathingsoap','shampoo','moisturizer']])
plt.title('Product Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sales')
plt.legend(title='Products', labels=['facecream', 'facewash', 'toothpaste', 'bathingsoap','shampoo','moisturizer'])
plt.grid(True)
plt.subplot(2, 2, 3)
data.drop(['month_number'], axis=1).plot(kind='bar', ax=plt.gca())
plt.title('Bar Chart for All Features/Attributes')
plt.xlabel('Month Number')
plt.ylabel('Values')
plt.xticks(rotation=0)
plt.grid(True)

plt.subplot(2, 2, 4).axis('off')
plt.tight_layout()
plt.show()
```
